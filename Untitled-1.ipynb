## Fase 2: Preparación para Clustering

Las variables categóricas fueron codificadas y los datos fueron normalizados para facilitar el uso de algoritmos de segmentación. Se aplicó PCA para una visualización en 2D.


```python
import pandas as pd
print("✅ pandas funciona correctamente")
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Codificación de variables categóricas
df_encoded = df.copy()
df_encoded['Horario_preferido'] = df_encoded['Horario_preferido'].map({'Mañana': 0, 'Tarde': 1, 'Noche': 2})
df_encoded['Tiene_espacio_estudio'] = df_encoded['Tiene_espacio_estudio'].map({'No': 0, 'Sí': 1})
df_encoded['Frecuencia_repaso_apuntes'] = df_encoded['Frecuencia_repaso_apuntes'].map({
    'Nunca': 0, 'Rara vez': 1, 'A veces': 2, 'Frecuentemente': 3, 'Siempre': 4
})
df_encoded['Usa_tecnicas_estudio'] = df_encoded['Usa_tecnicas_estudio'].map({'No': 0, 'Sí': 1})

# Selección de variables
features = [
    'Horas_de_estudio_por_semana', 'Horario_preferido', 'Tiene_espacio_estudio',
    'Frecuencia_repaso_apuntes', 'Usa_tecnicas_estudio', 'Calificacion_promedio'
]
X = df_encoded[features]

# Normalización
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Visualización previa
plt.figure(figsize=(8, 6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1])
plt.title("Visualización con PCA")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.tight_layout()
plt.show()
```

## Fase 3: Segmentación de Estudiantes

Se utilizó el algoritmo K-Means para agrupar estudiantes con hábitos similares. Esto permitió identificar perfiles asociados a calificaciones bajas.


```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Determinar número óptimo de clusters
inertia = []
silhouette_scores = []
K = range(2, 8)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(X_scaled, kmeans.labels_))

# Método del codo
plt.figure(figsize=(8, 4))
plt.plot(K, inertia, marker='o')
plt.title("Método del Codo")
plt.xlabel("Número de Clusters")
plt.ylabel("Inercia")
plt.tight_layout()
plt.show()

# Clustering final
k_optimo = 3
kmeans_final = KMeans(n_clusters=k_optimo, random_state=42)
df['Cluster'] = kmeans_final.fit_predict(X_scaled)

# Visualización final
plt.figure(figsize=(8, 6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=df['Cluster'], palette='Set2')
plt.title("Clusters Estudiantiles")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.tight_layout()
plt.show()
```

## Fase 4: Interpretación de Resultados

Se interpretó cada clúster generado y se extrajeron conclusiones en base a sus características promedio. Esto permite detectar qué tipo de estudiantes están en riesgo académico.


```python
# Agrupación y resumen solo con columnas numéricas
cluster_profiles = df.groupby("Cluster")[[
    "Horas_de_estudio_por_semana",
    "Calificacion_promedio"
]].mean().round(2)

print("Perfiles por Clúster:")
print(cluster_profiles)

# Interpretación automática
print("\nInterpretación de perfiles:\n")
for cluster_id, row in cluster_profiles.iterrows():
    print(f"🔹 Cluster {cluster_id}:")
    horas = row['Horas_de_estudio_por_semana']
    calificacion = row['Calificacion_promedio']
    
    if horas >= 8 and calificacion < 75:
        tipo = "Estudiantes esforzados pero con bajo rendimiento"
        recomendacion = "Revisar técnicas de estudio, apoyo emocional o estrategias cognitivas."
    elif horas < 5 and calificacion < 75:
        tipo = "Estudiantes desmotivados o poco comprometidos"
        recomendacion = "Aplicar talleres de motivación, técnicas de estudio y mentorías."
    elif horas < 7 and calificacion >= 85:
        tipo = "Estudiantes eficientes"
        recomendacion = "Pueden ser guías o mentores para otros."
    else:
        tipo = "Estudiantes promedio"
        recomendacion = "Pueden mejorar con intervención leve o seguimiento docente."

    print(f"   Tipo: {tipo}")
    print(f"   Recomendación: {recomendacion}\n")

```
